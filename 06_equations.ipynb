```python
# one equation
# one unknown
# and one solution

# x^0.5 = 2.5

from scipy.optimize import fsolve
import numpy as np

def model(x):
    eq = np.sqrt(x) - 2.5
    return eq

guess = 1
x0, = fsolve(model, guess)
print("solution:", x0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

y0 = 2.5
x = np.linspace(0,11,100)
y1 = np.sqrt(x)
y2 = np.full(100,y0) # y1 = 2.5
plt.plot(x, y1)
plt.plot(x, y2)

plt.plot(x0,y0,"ro")
```

```python
# one equation
# one unknown
# >1 solutions

# (x+2)^2 = 10

from scipy.optimize import fsolve
import numpy as np

def model(x):
    eq = (x+2)**2 - 10
    return eq

guess = 10
x0, = fsolve(model, guess)
print("solution 1:", x0)

guess = -10
x0, = fsolve(model, guess)
print("solution 2:",x0)
```

```python
# two equations
# two unknowns

from scipy.optimize import fsolve
import numpy as np

def system( unknowns ):
    x = unknowns[0] # unpack x
    y = unknowns[1] # unpack y
    
    eq1 = y - x**2 - 2*x # calculate eq1 for given x,y
    eq2 = y - 12 + x**2 # calculate eq2 for given x,y
    return [eq1, eq2] # return the results (0?)

# first guess pair
guess = [1, 1] # two unknowns = two guess; in order: x, y

x0, y0 = fsolve(system, guess)
print("x0, y0 = {} {}".format(x0, y0))

guess = [-10, -10] # different pair of guess values
x1, y1 = fsolve(system, guess)
print("x1, y1 = {} {}".format(x1, y1))

```
