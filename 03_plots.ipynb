```python
# plot a datapoints

x=np.array([1,2,3,4])
y=np.array([4,6,7,8])

plt.xlabel('X')
plt.ylabel('Y')
plt.title('A title')
plt.plot(x,y,'bo-')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# plot a function

x = np.linspace(0,60)

y1 = 3*x+10
y2 = 0.1*x**2 - 3*x + 10

plt.plot(x,y1,'b-', label='y1')
plt.plot(x,y2,'r--', label='y2')

plt.xlabel('x')
plt.ylabel('y1, y2')

plt.xlim(-100,100) # x range
plt.ylim(-50,300) # y range

plt.legend(loc='upper left') # loc = location
```
