```python
# komentarz

# wyrażenia arytmetyczne
print(2+2*23**2)
print((12+1.2)/232**3)


# zmienne

a = 4 # mg/mL <- jednostka jako komentarz
b = 4.562871
c = 3.4e-1 # =3.4x10^-1 = 0.34 <- notacja naukowa


print(a,b,c)
```

```python
# zmienne - działania

d = a+b+c
print(d)
```

```python
# formatowanie liczb przy wyświetlaniu wyników:

print("{:.2f}".format(d))

print("{:.4f}".format(d))

print("{:.2e}".format(d))

# to samo w jednej linijce
print("{:.2f} {:.4f} {:.2e}".format(d,d,d))
```
