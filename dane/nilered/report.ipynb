```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
tmac = np.genfromtxt('tmac-ia-nr.csv', delimiter=' ')
tmacw = np.genfromtxt('tmac-ia-w-nr.csv', delimiter=' ')

teac = np.genfromtxt('teac-ia-nr.csv', delimiter=' ')
teacw = np.genfromtxt('teac-ia-w-nr.csv', delimiter=' ')

cc = np.genfromtxt('cc-ia-nr.csv', delimiter=' ')
ccw = np.genfromtxt('cc-ia-w-nr.csv', delimiter=' ')

btmac = np.genfromtxt('btmac-ia-nr.csv', delimiter=' ')
btmacw = np.genfromtxt('btmac-ia-w-nr.csv', delimiter=' ')

bet = np.genfromtxt('bet-ia-nr.csv', delimiter=' ')
betw = np.genfromtxt('bet-ia-w-nr.csv', delimiter=' ')
```

```python
def difff(x,y):
    dx = x[1]-x[0]
    dydx = np.gradient(y, dx)
    return dydx
```

```python
plt.plot(tmac[:,0], tmac[:,1]/2-0.03, 'c-', label='tmac')
plt.plot(tmacw[:,0], tmacw[:,1], 'b--',label='tmac + w')
plt.legend()
plt.xlabel('Wavelength, nm')
plt.show()
plt.plot(teac[:,0], teac[:,1], 'c-', label='teac')
plt.plot(teacw[:,0], teacw[:,1]/2.3, 'b--',label='teac + w')
plt.legend()
plt.xlabel('Wavelength, nm')
plt.show()
plt.plot(cc[:,0], cc[:,1], 'c-', label='cc')
plt.plot(ccw[:,0], ccw[:,1]/2.3, 'b--',label='cc + w')
plt.legend()
plt.xlabel('Wavelength, nm')
plt.show()
plt.plot(btmac[:,0], btmac[:,1], 'c-', label='btmac')
plt.plot(btmacw[:,0], btmacw[:,1], 'b--',label='btmac + w')
plt.legend()
plt.xlabel('Wavelength, nm')
plt.show()
plt.plot(bet[:,0], bet[:,1], 'c-', label='bet')
plt.plot(betw[:,0], betw[:,1]/3, 'b--',label='bet + w')
plt.legend()
plt.xlabel('Wavelength, nm')
plt.show()
```

```python
plt.plot(cc[:,0], cc[:,1]/np.max(cc[:,1]), 'c-', label='cc')
plt.plot(teac[:,0], teac[:,1]/np.max(teac[:,1]), 'b-', label='teac')
plt.plot(tmac[:,0], (tmac[:,1]-0.05)/np.max(tmac[:,1]-0.05), 'g-', label='tmac')
plt.plot(btmac[:,0], btmac[:,1]/np.max(btmac[:,1]), 'y--', label='btmac')
plt.plot(bet[:,0], (bet[:,1]-0.12)/np.max(bet[0:16*9,1]-0.12), 'm-', label='bet')
plt.xlim(450,750)
plt.ylim(0,2)
plt.legend(ncol=3)
plt.xlabel('Wavelength, nm')
plt.savefig('all.pdf')
plt.savefig('all.png', dpi=300)
```
